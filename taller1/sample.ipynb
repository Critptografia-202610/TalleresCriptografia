### 1.a - valores posibles de a y b

Como $E_{a,b}(x) = (a \cdot x + b) \bmod 26$, entonces $a$ y $b$ pueden tomar cualquier valor en $\mathbb{Z}_{26}$, es decir, $a,b \in {0..25}$. Esto se debe a que el resultado de la función de cifrado se reduce módulo 26, lo que significa que cualquier valor de $a$ y $b$ fuera de este rango se reducirá a un valor dentro del rango.

### 1.b - claves válidas

Sabiendo que las claves se definen como $K=(a,b)$, es necesario que $a$ sea invertible en $\mathbb{Z}_{26}$, es decir, que $a$ sea coprimo con 26. Por otro lado, también es necesario que $b \in \mathbb{Z}_{26}$. Por lo tanto:

```python
# Encontrar todos los co-primos de 26 \in [0..25]
import math # para usar la función gcd
a_validos = [x for x in range(26) if math.gcd(x, 26) == 1]

print(f"Co-primos totales de 26: {len(a_validos)}")
print("Valores posibles de a:")
for a in a_validos:
    print(f"a = {a}")
```

```python
# Encontrar todos los valores posibles de b \in [0..25]
b_validos = [x for x in range(26)]

print("Valores posibles de b:")
for b in b_validos:
    print(f"b = {b}")
```

```python
# Encontrar todas las claves válidas K = (a, b)
claves_validas = []

# Matemáticamente, número de claves válidas = número de a válidos * número de b válidos = 12 * 26 = 312
for a in a_validos:
    for b in b_validos:
        claves_validas.append((a, b))

# Verificar que el número de claves válidas coincide con el producto del número de a válidos y b válidos
assert len(claves_validas) == len(a_validos) * 26, "El número de claves válidas no coincide con el producto del número de a válidos y b válidos"

print(f"Número de claves válidas (a, b): {len(claves_validas)}")
print(f"Claves válidas:")
for clave in claves_validas:
    print(f"K = {clave}")
```
